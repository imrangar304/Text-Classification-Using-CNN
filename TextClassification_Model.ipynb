```python
embedding_dict = {}

with open('/content/drive/MyDrive/NaiveBayesAssignment/glove_vectors', 'rb') as f:
    model = pickle.load(f)
    glove_words =  set(model.keys())
    for word in glove_words:
        embedding_dict[word] = model[word]
    f.close()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model 
from tensorflow.keras.layers import Input,Dense,Conv1D,Flatten,Embedding,MaxPool1D,concatenate,Dropout
from tensorflow.keras.callbacks import ModelCheckpoint,TensorBoard,EarlyStopping
from tensorflow.keras.optimizers import Adam


input_layer = Input(shape=(maxlen))

embed = embedding_layer(input_layer)

#tf.keras.layers.Conv1D(
#    filters =	Integer, the dimensionality of the output space (i.e. the number of output filters in the convolution
#    kernel_size = An integer or tuple/list of a single integer, specifying the length of the 1D convolution window. 
#    activation=relu,
#    kernel_initializer=he_normal,
#    kernel_regularizer='tf.keras.regularizers.l2'
#)

Conv1D_m = Conv1D(25,4,activation="relu",kernel_initializer =tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(embed)
Conv1D_n = Conv1D(30,4,activation="relu",kernel_initializer =tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(embed)
Conv1D_o = Conv1D(35,4,activation="relu",kernel_initializer = tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(embed)

second_layer = concatenate([Conv1D_m,Conv1D_n,Conv1D_o])

max_pool_1 = MaxPool1D(3)(second_layer)

Conv1D_i = Conv1D(25,3,activation="relu",kernel_initializer = tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(max_pool_1)
Conv1D_j = Conv1D(30,3,activation="relu",kernel_initializer = tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(max_pool_1)
Conv1D_k =  Conv1D(35,3,activation="relu",kernel_initializer = tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(max_pool_1)

third_layer = concatenate([Conv1D_i,Conv1D_j,Conv1D_k])

max_pool_2 = MaxPool1D(3)(third_layer)

fourth_layer = Conv1D(30,3,activation='relu',kernel_initializer = tf.keras.initializers.he_normal(seed=42),kernel_regularizer=tf.keras.regularizers.l2())(max_pool_2)

flatten = Flatten()(fourth_layer)

dropout_layer = Dropout(0.2)(flatten)

dense_layer = Dense(64,activation="relu",kernel_initializer = tf.keras.initializers.he_normal())(dropout_layer)
              
output_layer = Dense(20,activation="softmax",kernel_initializer= tf.keras.initializers.glorot_normal())(dense_layer)
              

model =Model(inputs=input_layer,outputs=output_layer) 
             
        
```
